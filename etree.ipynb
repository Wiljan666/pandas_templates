```python
#import pandas as pd
import xml.etree.ElementTree as ET
import io



def iter_docs(etree):
    d={}
    root=etree.getroot()
    d['root']=root.tag
    print(root.tag)
    for child in root:
        print(child.tag)
        d['customer']=child.tag
        for c in child:
            print(c.tag)
            d['test']=c.tag
            for c1 in c:
                d['platform']=c1.tag
                print(c1.tag)
                for c2 in c1:
                    passed = c2.find('passed').text
                    d['passed']=passed
                    print(passed)
                    failed = c2.find('failed').text
                    print(failed)
                    d['failed']=failed
                    total = c2.find('total').text
                    print(total)
                    d['total']=total
                    logdir = c2.find('logdir').text
                    print(logdir)
                    d['logdir']=logdir
                    starttime=c2.find('starttime').text
                    print(starttime)
                    d['starttime']=starttime
                    finishtime=c2.find('finishtime').text
                    print(finishtime)
                    d['finishtime']=finishtime
                    killed=c2.find('killed').text
                    print(killed) 
                    d['killed']=killed
    print(d)
xml_data = io.StringIO(u'''\
<stream>
<ams>
<CLASSIC>
<lnx86>
<amss>
<failed>1</failed>
<finishtime>03-Sep-2020.09:32:24</finishtime>
<gmatrix>1</gmatrix>
<killed>0</killed>
<logdir>/lan/csv/amscm_s4/ws/amssim/main_ams_sj/mmsim_1910_isr/amss_lnx86/log/09-03-2020</logdir>
<lost>1</lost>
<machine>sjfhua019</machine>
<outdir>/lan/csv/amscm_s4/ws/amssim/main_ams_sj/mmsim_1910_isr/amss_lnx86/out/09-03-2020/</outdir>
<passed>3605</passed>
<passrate>99</passrate>
<starttime>03-Sep-2020.08:40:02</starttime>
<status>finished</status>
<syserr>0</syserr>
<timezone>PDT</timezone>
<total>3606</total>
<use_def_test_output>1</use_def_test_output>
</amss>
</lnx86>
</CLASSIC>
</ams>
</stream>
''')


etree = ET.parse(xml_data)
iter_docs(etree)
```
